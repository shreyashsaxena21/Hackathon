# Top ten and bottom ten countries analysis on basis happiness score

```python
top_bottom_10 = df.nlargest(10, 'Ladder score').append(df.nsmallest(10, 'Ladder score'))
plt.figure(figsize=(10, 6))
sns.barplot(x='Ladder score', y='Country name', data=top_bottom_10, palette='viridis')
plt.xlabel('Happiness Score')
plt.title('Top 10 and Bottom 10 Countries by Happiness Score (2021)')
plt.show()
```

# Happiest and Unhappiest Countries in 2021,2022,2023

```python
happiness_years = ['2021', '2022', '2023']

for year in happiness_years:
    # Filter the data for the specified year
    year_data = df[df['year'] == year]
    
    # Check if there are data points for the specified year
    if not year_data.empty:
        # Get the top 10 and bottom 10 countries by happiness score
        happiest = year_data.nlargest(10, 'Ladder score')
        unhappiest = year_data.nsmallest(10, 'Ladder score')
        
        # Create bar plots for the happiest and unhappiest countries
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Ladder score', y='Country name', data=happiest, palette='coolwarm')
        plt.xlabel('Happiness Score')
        plt.title(f'Happiest Countries in {year}')
        plt.show()
    else:
        print(f"No data available for the year {year}")
```
